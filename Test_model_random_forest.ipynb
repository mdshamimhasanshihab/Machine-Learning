```python
import pandas as pd
df=pd.read_csv("train.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_gender=LabelEncoder()
le_s11=LabelEncoder()
le_s12=LabelEncoder()
le_s16=LabelEncoder()
le_s17=LabelEncoder()
le_s18=LabelEncoder()
le_s48=LabelEncoder()
le_s52=LabelEncoder()
le_s54=LabelEncoder()
le_s55=LabelEncoder()
le_s58=LabelEncoder()
le_s69=LabelEncoder()
le_s70=LabelEncoder()
le_s71=LabelEncoder()
le_s53=LabelEncoder()
le_s56=LabelEncoder()
le_s57=LabelEncoder()
le_s59=LabelEncoder()
df['s11_n']=le_gender.fit_transform(df['gender'])
df['S12_n']=le_s12.fit_transform(df['s12'])
df['s16_n']=le_s16.fit_transform(df['s16'])
df['s17_n']=le_s17.fit_transform(df['s17'])
df['s18_n']=le_s18.fit_transform(df['s18'])
df['s48_n']=le_s48.fit_transform(df['s48'])
df['s52_n']=le_s52.fit_transform(df['s52'])
for c in  df.columns:
    df[c].fillna('No', inplace = True)
df['s53_n']=le_s53.fit_transform(df['s53'])
df['s54_n']=le_s54.fit_transform(df['s54'])
df['s55_n']=le_s55.fit_transform(df['s55'])
df['s56_n']=le_s56.fit_transform(df['s56'])
df['s57_n']=le_s57.fit_transform(df['s57'])
df['s58_n']=le_s58.fit_transform(df['s58'])
df['s59_n']=le_s59.fit_transform(df['s59'])
df['s69_n']=le_s69.fit_transform(df['s69'])
df['s70_n']=le_s70.fit_transform(df['s70'])
df['s71_n']=le_s71.fit_transform(df['s71'])
```

```python
target=df['label']
input_n=df.drop(['id','gender','s11','s12','s16','s17','s18','s52','s53','s54','s55','s56','s57','s58','s59','s69','s70','s71'],axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input_n,target,test_size=0.30)
from sklearn.ensemble import RandomForestClassifier 
model=RandomForestClassifier(n_estimators=20)
```

```python
X_train
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
