```python
import pandas as pd
```

```python
df=pd.read_excel("Titanic.xlsx")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex=LabelEncoder()
inputs=df
```

```python
inputs['Sex_n']=le_Sex.fit_transform(inputs['Sex'])
inputs
```

```python
input_n=inputs.drop(['PassengerId','Sex','SibSp','Parch','Ticket','Cabin','Embarked','Name'],axis='columns')
input_n
```

```python
input_n=inputs.drop(['PassengerId','Sex','SibSp','Parch','Ticket','Cabin','Embarked','Name'],axis='columns')
input_n
```

```python
nullfree=input_n.dropna()
nullfree
```

```python
train_info=nullfree.drop('Survived',axis='columns')
target=nullfree['Survived']
target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_info,target,test_size=0.25)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
labels = ["{0}".format(i, i + 5) for i in range(0, 100, 5)]
X_train["Age_n"] = pd.cut(X_train.Age, range(0, 105, 5), right=False, labels=labels)
X_train
```

```python
X_train['Fare'].max()
```

```python
labels1 = ["{0}".format(j, j + 9) for j in range(0, 515, 5)]
X_train["Fare_n"] = pd.cut(X_train.Fare, range(0, 520, 5), right=False, labels=labels1)
X_train
```

```python
X_train=X_train.drop(['Age','Fare'],axis='columns')
X_train
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
